```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("ecommerce.csv")
df.head()
```

```python
data = {
    "Order_ID": range(1,501),
    "Order_Date": pd.date_range(start="2023-01-01", periods=500, freq="D"),
    "Region": np.random.choice(["North","South","East","West"], 500),
    "Category": np.random.choice(["Electronics","Clothing","Home","Sports"], 500),
    "Sales": np.random.randint(100, 5000, 500),
    "Profit": np.random.randint(-500, 1500, 500),
    "Returned": np.random.choice(["Yes","No"], 500, p=[0.1,0.9])
}

df = pd.DataFrame(data)
df.head()
```

```python
df.isnull().sum()
df.duplicated().sum()
df.drop_duplicates(inplace=True)
```

```python
df["Profit_Margin"] = (df["Profit"] / df["Sales"]) * 100
df
```
